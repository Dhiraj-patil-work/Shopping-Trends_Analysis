```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Figure size
plt.figure(figsize=(12, 6))

# Distribution of Purchase Amount
sns.histplot(df["Purchase Amount (USD)"], bins=20, kde=True, color="blue")
plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")
plt.show()
```

**This histogram shows the distribution of purchase amounts, with most transactions concentrated around $40$80.**

```python
# Bar plot: Average Purchase Amount by Gender
plt.figure(figsize=(8, 5))
sns.barplot(x="Gender", y="Purchase Amount (USD)", data=df, estimator=lambda x: x.mean(), palette="coolwarm")
plt.title("Average Purchase Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Purchase Amount (USD)")
plt.show()
```

**This bar chart shows the average purchase amount by gender. If there's a noticeable difference, it could indicate varying spending habits between male and female customers.**

```python
# Bar plot: Average Purchase Amount by Category
plt.figure(figsize=(10, 5))
sns.barplot(x="Category", y="Purchase Amount (USD)", data=df, estimator=lambda x: x.mean(), palette="viridis")
plt.title("Average Purchase Amount by Category")
plt.xlabel("Category")
plt.ylabel("Average Purchase Amount (USD)")
plt.xticks(rotation=45)
plt.show()
```

**This chart shows the average purchase amount across different product categories.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style
sns.set_style("whitegrid")
```

```python
# Histogram of Age
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=20, kde=True, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram of Purchase Amount
plt.figure(figsize=(8, 5))
sns.histplot(df["Purchase Amount (USD)"], bins=20, kde=True, color='green')
plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot of Review Ratings
plt.figure(figsize=(6, 4))
sns.boxplot(y=df["Review Rating"], color='orange')
plt.title("Review Rating Distribution")
plt.ylabel("Review Rating")
plt.show()
```

```python
# Count plots for categorical variables
categorical_features = ['Gender', 'Category', 'Season', 'Subscription Status', 'Payment Method']

plt.figure(figsize=(15, 10))
for i, col in enumerate(categorical_features):
    plt.subplot(2, 3, i+1)
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette='coolwarm')
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
# Boxplot of Purchase Amount by Gender
plt.figure(figsize=(6, 4))
sns.boxplot(x='Gender', y='Purchase Amount (USD)', data=df, palette="coolwarm")
plt.title("Purchase Amount by Gender")
plt.show()
```

```python
# Purchase Amount vs. Subscription Status
plt.figure(figsize=(6, 4))
sns.boxplot(x='Subscription Status', y='Purchase Amount (USD)', data=df, palette="coolwarm")
plt.title("Purchase Amount by Subscription Status")
plt.show()
```

```python
numerical_df = df.select_dtypes(include=['number'])  # Selects only numerical columns
```

```python
corr_matrix = numerical_df.corr()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

Key Insights-
1. Customer Demographics (Univariate Analysis)
Age Distribution: Most customers fall within a certain age range (e.g., 25-40). This can help in targeted marketing.
Gender Distribution: If theres a skew (e.g., more females than males), companies can create gender-focused campaigns.
2. Purchase Behavior (Univariate Analysis)
Purchase Amount Distribution: If its right-skewed, a small group of high spenders contributes significantly to revenue.
Seasonal Trends: If more purchases occur in winter, marketing campaigns should focus on winter collections.
3. Impact of Gender on Purchases (Bivariate Analysis)
If males or females have higher average purchase amounts, companies can tailor discounts and offers.
4. Subscription Status & Spending Patterns
If subscribed customers spend more, the company should promote subscription benefits.
If non-subscribed customers spend more, the subscription model may need improvement.
5. Preferred Payment Methods
If one payment method dominates, optimizing checkout experiences for that method can improve conversions.
6. Correlation Between Variables
If previous purchases strongly correlate with higher spending, loyalty programs can be enhanced to boost retention. **bold text**
